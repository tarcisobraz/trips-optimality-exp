```python
import pandas as pd
```

```python
buste_data = pd.read_csv('/local/tarciso/masters/data/bus_trips/test/buste-output-latest/2017_05_09_veiculos/part-00000', dtype = {'route': str})
```

```python
buste_data.head()
```

```python
num_trips_per_card = buste_data.cardNum.value_counts().reset_index() \
                        .rename(index=str, columns={'index':'cardNum','cardNum':'num_trips'}) \
                        .drop(['0'])
```

```python
type(num_trips_per_card)
```

```python
num_trips_per_card
```

```python
filtered_cards = num_trips_per_card[(num_trips_per_card['num_trips'] >= 2) & (num_trips_per_card['num_trips'] <= 4)]
```

```python
selected_cards = filtered_cards.sample(n=20, random_state=1229)
```

```python
selected_cards
```

```python
buste_sample = buste_data[buste_data['cardNum'].isin(selected_cards['cardNum'])] \
                    .sort_values(['cardNum','cardTimestamp'])
```
